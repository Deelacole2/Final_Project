```python
# Dependecy import
import pandas as pd
import matplotlib.pyplot as plt

# File to Load (Remember to change the path if needed.)
data_to_load = "Resources/us_county_sociohealth_data.csv"

sociohealth_df = pd.read_csv(data_to_load)

sociohealth_df.head()
```

```python
# Basic EDA to see what we have
print(sociohealth_df.shape)
sociohealth_df.columns
```

```python
sociohealth_df.dtypes
```

```python
sociohealth_df.describe()
```

```python
state_array = sociohealth_df['state'].unique()
state_array
```

```python
# Average county population for each state

Avg_pop_per_state = sociohealth_df.groupby(["state"]).mean()['total_population']
Avg_pop_per_state
```

```python
# Getting the average_number_of_physically_unhealthy_days averaging the column averages grouped by state

Avg_number_of_unhealthy_days = sociohealth_df.groupby(["state"]).mean()['average_number_of_physically_unhealthy_days']
Avg_number_of_unhealthy_days
```

```python


Avg_number_of_uninsured = sociohealth_df.groupby(["state"]).mean()['num_uninsured']
Avg_number_of_uninsured
```

```python
state_metric_df = pd.DataFrame({
    "Average Population for Each County by State" : Avg_pop_per_state,
    "Average Uninsured Per State" : Avg_number_of_uninsured,
    "Average Number of Unhealthy Days" : Avg_number_of_unhealthy_days
    })

state_metric_df
```

```python
state_metric_df["Average Uninsured Per State"] = state_metric_df["Average Uninsured Per State"].map("{:.2f}".format)
state_metric_df["Average Number of Unhealthy Days"] = state_metric_df["Average Number of Unhealthy Days"].map("{:.2f}".format)
state_metric_df["Average Population for Each County by State"] = state_metric_df["Average Population for Each County by State"].map("{:.2f}".format)
state_metric_df
```
